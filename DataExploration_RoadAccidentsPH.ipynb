##Road Accident Statistics

```python
accident_by_year = ar_df[["year", "total_accidents"]]
year = ar_df["year"]
total_accident = ar_df["total_accidents"]
accident_by_year
```

```python
plt.figure(figsize=(15,10))
plt.xticks( fontsize=16)
plt.yticks( fontsize=16)
plt.plot(year, total_accident)
plt.title("Road Accident Statistical Trend", fontsize=20)
plt.ylabel("Total Number of Accidents", fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
car_involved_yearly= ar_df[["year", "car_involved"]]
year = ar_df["year"]
number_of_car_involved = ar_df["car_involved"]
car_involved_yearly
```

```python
plt.figure(figsize=(15,10))
plt.xticks( fontsize=16)
plt.yticks( fontsize=16)
plt.plot(year, number_of_car_involved)
plt.title("Cars Involved in Accident Statistical Trend", fontsize=20)
plt.ylabel("Total Number of Cars Involved", fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.xticks( fontsize=16)
plt.yticks( fontsize=16)
plt.plot(year, number_of_car_involved)
plt.plot(year, total_accident)
plt.title("Total Accidents and the Cars Involved Statistical Trend", fontsize=20)
plt.ylabel("Total Number of Cars Involved", fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.legend(["Total Number of Accidents", "Total Number of Cars Involved"], fontsize=20)
plt.show()
```

```python
ave_killed_d = {"type": ["drivers_killed", "passengers_killed", "pedestrians_killed"], "mean":[ar_df["drivers_killed"].mean(), ar_df["passengers_killed"].mean(), ar_df["pedestrians_killed"].mean()]}
ave_killed_df = pd.DataFrame(ave_killed_d)
```

```python
ave_killed_newdf = ave_killed_df.set_index("type")
ave_killed_newdf
```

```python
ave_killed_newdf.plot(kind="bar", figsize=(15,10), fontsize = 20, rot=45)
plt.title("Average deaths from 2005-2020", fontsize=20)
plt.xlabel("")
```

```python
rm_df = pd.read_csv("report_by_month.csv", index_col = "year")
rm_df.head()
```
